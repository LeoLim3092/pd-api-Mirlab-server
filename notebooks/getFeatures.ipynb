```python
import os
import pandas as pd
import json
import numpy as np
import re
hand_features_name = ["Right Tapping Time", "Right Tapping Time Change", "Right Tapping Distance", "Right Tapping Frequency",
                      "Right Tapping Intensity", "Right Tapping Power", "Right Tapping Frequency Change", "Left Tapping Time",
                      "Left Tapping Time Change", "Left Tapping Distance", "Left Tapping Frequency", 
                      "Left Tapping Intensity", "Left Tapping Power", "Left Tapping Frequency Change"]
```

```python
with open("/mnt/other/hand_prediction_proba.json", 'r') as f:
    data = json.load(f)
```

```python
f_dt = {}
p = r"(\d{8})_(\d+)"
for k, v in data.items():

    hand_f_pt = f"/mnt/other/results/{k}/hand_feature.npy"
    hf = np.load(hand_f_pt)
    match = re.match(p, k)
    if match:
        date = match.group(1)
        pid = int(match.group(2))
    f_dt[(date, pid)] = hf

```

```python
df = pd.DataFrame.from_dict(f_dt).T
df.columns = hand_features_name
df.to_csv("/mnt/other/20260316_previous_hand_features.csv")
```
